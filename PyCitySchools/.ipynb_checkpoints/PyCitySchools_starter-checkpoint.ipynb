```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#School summary - group school_data_complete by school_name
grp_school_name_df = school_data_complete.groupby(['school_name'])

#school total students
school_tot_students = grp_school_name_df['Student ID'].count()


#school per student budget
school_per_student_budget = grp_school_name_df['budget'].unique()/school_tot_students
school_per_student_budget



#Adding variables school_tot_students and school_per_student_budget
school_data['Per Student Budget'] = [i[0] for i in school_per_student_budget]
school_data['Total Students'] = school_tot_students
school_data.head()

#average math score
school_avg_math_score = grp_school_name_df['math_score'].mean()
school_avg_math_score


#average reading score
school_avg_reading_score = grp_school_name_df['reading_score'].mean()
school_avg_reading_score


#passing math
grp_school_pass_math = passing_math.groupby(['school_name'])
school_pass_math_percent = grp_school_pass_math['Student ID'].count()/school_data['Total Students']*100
#print(school_pass_math_percent)


#passing reading
grp_school_pass_reading = passing_reading.groupby(['school_name'])
school_pass_reading_percent = grp_school_pass_reading['Student ID'].count()/school_data['Total Students']*100
#print(school_pass_reading_percent)

#overall passing rate
school_overall_pass_rate = (school_pass_math_percent + school_pass_reading_percent)/2


#create new dataframe
school_summ_df = pd.DataFrame({"School Type":school_data['type'],
                              "Total Students":school_data['Total Students'],
                               "Total School Budget":school_data['budget'],
                               "Per Student Budget":school_data['Per Student Budget'],
                               "Average Math Score":school_avg_math_score,
                               "Average Reading Score":school_avg_reading_score,
                               "% Passing Math":school_pass_math_percent,
                               "% Passing Reading":school_pass_reading_percent,
                               "% Overall Passing Rate": school_overall_pass_rate
                              })

#format the data values
school_summ_df['Total Students'] = school_summ_df['Total Students'].map("{:,}".format)
school_summ_df['Total School Budget'] = school_summ_df['Total School Budget'].map("${:,.2f}".format)
school_summ_df['Per Student Budget'] = school_summ_df['Per Student Budget'].map("${:,.2f}".format)



school_summ_df.sort_values(by=["% Overall Passing Rate"], ascending=False).head()
```
